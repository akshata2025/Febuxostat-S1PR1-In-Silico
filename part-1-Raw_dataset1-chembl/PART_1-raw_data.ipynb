```python
import pandas as pd
data = pd.read_csv('/content/DOWNLOAD-bMvTskxhFg6a9hkRROdQerj2G5k0QhOkPXtZhAszrPs_eq_.csv', delimiter=';')
data.head()
```

```python
# Number of rows
len(data)
```

```python
# Check columns
data.columns
```

```python
# Keep only relevant columns
df = data[['Molecule ChEMBL ID', 'Smiles', 'Standard Value', 'Standard Type', 'pChEMBL Value']]
```

```python
## check null , duplicates values
df.isnull().sum()
```

```python
# Drop duplicates
df = df.drop_duplicates()
```

```python
# Drop rows where pChEMBL Value is missing
df = df.dropna(subset=['pChEMBL Value'])
```

```python
# Reset index
df = df.reset_index(drop=True)
```

```python
# Classify compounds
# Here we assume pChEMBL >= 6 is Active, < 6 is Inactive
df['Activity_Class'] = df['pChEMBL Value'].apply(lambda x: 'Active' if x >= 6 else 'Inactive')
```

```python

# Check class distribution
print(df['Activity_Class'].value_counts())
```

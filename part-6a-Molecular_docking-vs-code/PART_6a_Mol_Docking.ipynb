```python
import os
import subprocess
import csv
```

```python
# Vina executable full path
vina_path = "/opt/homebrew/bin/vina"  # full path to Vina
```

```python
# Paths
receptor = "cleaned_7EW0.pdbqt"
ligand_folder = "Ligand_PDBQT"
output_folder = "docking_results"
os.makedirs(output_folder, exist_ok=True)
```

```python
# Grid box parameters
center_x, center_y, center_z = 107.075, 112.817, 74.658
size_x, size_y, size_z = 17.5, 7.0, 19.0
```

```python
# Get all ligand files
ligand_files = [f for f in os.listdir(ligand_folder) if f.endswith(".pdbqt")]
ligand_files.sort(key=lambda x: int(x.split("_")[0]))
ligand_files = ligand_files[:500]  # Only first 10 ligands test
```

```python

# CSV to save best affinities
csv_file = os.path.join(output_folder, "docking_results.csv")
with open(csv_file, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Ligand", "Best_Affinity"])
```

```python
# Docking loop
total = len(ligand_files)
for idx, ligand_file in enumerate(ligand_files, 1):
    ligand_path = os.path.join(ligand_folder, ligand_file)
    output_path = os.path.join(output_folder, f"{ligand_file[:-6]}_out.pdbqt")

    # Skip already docked ligands
    if os.path.exists(output_path):
        print(f"[{idx}/{total}] {ligand_file} already docked, skipping...")
        continue

    try:
        # Run Vina
        cmd = [
            vina_path,
            "--receptor", receptor,
            "--ligand", ligand_path,
            "--center_x", str(center_x),
            "--center_y", str(center_y),
            "--center_z", str(center_z),
            "--size_x", str(size_x),
            "--size_y", str(size_y),
            "--size_z", str(size_z),
            "--out", output_path,
            "--exhaustiveness", "8",
            "--num_modes", "5"
        ]
        subprocess.run(cmd, check=True)

        # Extract best affinity
        best_affinity = None
        with open(output_path, "r") as f_out:
            for line in f_out:
                if line.startswith("REMARK VINA RESULT:"):
                    best_affinity = float(line.split()[3])
                    break

        # If affinity not found, consider file corrupted
        if best_affinity is None:
            raise ValueError("No affinity found, possibly corrupted file.")

        # Save to CSV
        with open(csv_file, "a", newline="") as f:
            writer = csv.writer(f)
            writer.writerow([ligand_file, best_affinity])

        print(f"[{idx}/{total}] {ligand_file} docked, best affinity: {best_affinity}")

    except Exception as e:
        print(f"[{idx}/{total}] {ligand_file} skipped due to error: {e}")
        continue
```

```python
import os
import pandas as pd

# Folder where your docking results were saved
output_folder = "docking_results"  # change if different

# Collect results
results = []
for file in sorted(os.listdir(output_folder)):
    if file.endswith("_out.pdbqt"):
        ligand_name = file.replace("_out.pdbqt", "")
        filepath = os.path.join(output_folder, file)

        with open(filepath, "r") as f:
            lines = f.readlines()

        # Extract affinities from REMARK VINA RESULT lines
        affinities = []
        for line in lines:
            if line.strip().startswith("REMARK VINA RESULT"):
                parts = line.split()
                affinities.append(float(parts[3]))  # kcal/mol

        if affinities:
            best_affinity = min(affinities)  # most negative = best binding
            results.append([ligand_name, best_affinity])

# Convert to DataFrame
df_results = pd.DataFrame(results, columns=["Ligand", "Best_Affinity_(kcal/mol)"])

# Sort affinities: strongest binders first
df_results = df_results.sort_values(by="Best_Affinity_(kcal/mol)", ascending=True).reset_index(drop=True)

# Save results to CSV
df_results.to_csv(os.path.join(output_folder, "docking_summary.csv"), index=False)

# Display the summary
print("✅ Docking summary saved in docking_results/docking_summary.csv")
print(df_results)
```

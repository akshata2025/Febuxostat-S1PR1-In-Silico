```python
 #Prepare the ligand: generate 3D coordinates, add hydrogens, convert to PDBQT

 # generate 3D and add hydrogens (writes JEU.pdb)
!obabel JEU.sdf -O JEU.pdb --gen3d -h
```

```python
# convert pdb -> pdbqt (AutoDock Vina format)
!obabel JEU.pdb -O JEU.pdbqt
```

```python
## These are the grid cordinates already caLculated fron pymol

# center_x, center_y, center_z = 107.075, 112.817, 74.658
# size_x, size_y, size_z = 17.5, 7.0, 19.0
```

```python
# Step 3 – Create the Vina configuration file

%%bash
cat > config.txt << EOF
receptor = cleaned_7EW0.pdbqt
ligand = JEU.pdbqt

center_x = 107.075
center_y = 112.817
center_z = 74.658

size_x = 17.5
size_y = 7.0
size_z = 19.0

exhaustiveness = 8
num_modes = 5
out = 7EW0-JEU.pdbqt
EOF

echo "===== config.txt ====="
cat config.txt
```

```python
# Step 4 – Run AutoDock Vina

!vina --config config.txt
```
